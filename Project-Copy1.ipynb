#  Model Building

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV 
from sklearn.model_selection import StratifiedKFold
```

```python
# Split the data into training and test for model building and evaluation
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x, y, test_size=0.15, random_state=42, stratify= y)
```

#  Trying Under Sampling as model is not learning the pattern for Imbalanced Class

```python
#imbalance treatement required
import imblearn
from xgboost import XGBClassifier
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy=0.25, random_state = 42)
x_rus, y_rus = rus.fit_resample(x_train, y_train)
print("imbalance data", x_train.value_counts(),y_train.value_counts())
print("balanced data",x_rus.value_counts(), y_rus.value_counts())
```

```python
y_rus.value_counts(normalize=True)
```

# XGBoost with undersampling

```python
xgb = XGBClassifier(random_state = 1)
xgb.fit(x_rus,y_rus)
```

```python
y_pred_train = xgb.predict(x_rus)
y_pred_test = xgb.predict(x_test)
```

```python
print(confusion_matrix(y_rus, y_pred_train))
print("*******************"*3)
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y_rus, y_pred_train))
print("*******************"*3)
print(classification_report(y_test, y_pred_test))
```

```python
print("Training Accuracy", accuracy_score(y_rus, y_pred_train))
print("*******************"*3)
print("Test Accuracy",accuracy_score(y_test, y_pred_test))
```

```python
from sklearn.model_selection import cross_val_score

training_grid  = cross_val_score(xgb, x_rus, y_rus, cv=10)
test_grid = cross_val_score(xgb, x_test, y_test, cv=10)
```

```python
print("Training Accuracy", training_grid.mean())
print("*****************"*3)
print("Test Accuracy", test_grid.mean())
```

```python
training_grid  = cross_val_score(xgb, x_rus, y_rus, cv=10, scoring= 'f1')
test_grid = cross_val_score(xgb, x_test, y_test, cv=10, scoring = 'f1')
```

```python
print("Training Accuracy", training_grid.mean())
print("*****************"*3)
print("Test Accuracy", test_grid.mean())
```
