```python
x = final_df.drop(['Potability'], axis=1)
y = final_df['Potability']
```

```python
y
```

```python
# Feature Scaling required
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
x_scaler
```

```python
x = pd.DataFrame(x_scaler, columns = x.columns[:])
x.head(10)
```

```python
# imbalance treatement required
import imblearn
from imblearn.over_sampling import RandomOverSampler, SMOTE
ros = RandomOverSampler()
x_ros, y_ros = ros.fit_resample(x, y)
print("imbalance data", y.value_counts())
print("balanced data", y_ros.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV 
```

```python
# Split the data into training and test for model building and evaluation
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_ros, y_ros, test_size=0.20, random_state=42)
```

```python
rf = RandomForestClassifier(criterion='entropy', max_features='sqrt',
                       n_estimators=250, random_state=42)
rf.fit(x_train,y_train)
```

```python
y_grid_train = rf.predict(x_train)
y_grid_test = rf.predict(x_test)
```

```python
print(confusion_matrix(y_train, y_grid_train))
print("*******************"*3)
print(confusion_matrix(y_test, y_grid_test))
```

```python
print(classification_report(y_train, y_grid_train))
print("*******************"*3)
print(classification_report(y_test, y_grid_test))
```

```python
print("Training Accuracy", accuracy_score(y_train, y_grid_train))
print("*******************"*3)
print("Test Accuracy",accuracy_score(y_test, y_grid_test))
```

```python
from sklearn.model_selection import cross_val_score

training_grid  = cross_val_score(rf, x_train, y_train, cv=10)
test_grid = cross_val_score(rf, x_test, y_test, cv=10)
```
